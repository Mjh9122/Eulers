```python
# Solution to problem 72
# Sieve is copied, rest is my own

import math
import numpy as np
```

```python
def sieveOfAtkin(limit):
    P = [2,3]
    sieve=[False]*(limit+1)
    for x in np.arange(1,int(math.sqrt(limit))+1):
        for y in np.arange(1,int(math.sqrt(limit))+1):
            n = 4*x**2 + y**2
            if n<=limit and (n%12==1 or n%12==5) : sieve[n] = not sieve[n]
            n = 3*x**2+y**2
            if n<= limit and n%12==7 : sieve[n] = not sieve[n]
            n = 3*x**2 - y**2
            if x>y and n<=limit and n%12==11 : sieve[n] = not sieve[n]
    for x in np.arange(5,int(math.sqrt(limit))):
        if sieve[x]:
            for y in np.arange(x**2,limit+1,x**2):
                sieve[y] = False
    for p in np.arange(5,limit):
        if sieve[p] : P.append(p)
    return P
```

```python
def check_single_primes(count, unchecked, primes, eliminated):
    for p in primes:
        to_check = [p]
        if p < 1_000_000 ** (1/2):
            while(to_check[-1] * p < 1_000_000):
                to_check.append(to_check[-1]*p)
        count += (1_000_000 - len(eliminated[p])) * len(to_check)
        for a in to_check:
            if unchecked[a] == 1:
                raise Exception
            else:
                unchecked[a] = 1
    print(sum(unchecked)) 
    return count, unchecked
```

```python
def check_double_primes(count, unchecked, primes, eliminated):
    for i, p1 in enumerate(primes):
        for p2 in primes[i+1::]:
            if p1 > 1_000_000/p2:
                break
            else:
                to_check = [p1*p2]
                while(1_000_000/to_check[-1] >= p1):
                    to_check.append(to_check[-1]*p1)
                stop = False
                while not stop:
                    stop = True
                    for e in to_check:
                        if 1_000_000/e >= p2:
                            if e * p2 not in to_check:
                                stop = False
                                to_check.append(e * p2)
            elims = eliminated[p1].union(eliminated[p2])
            count += (1_000_000-len(elims)) * len(to_check)
            for a in to_check:
                if unchecked[a] == 1:
                    raise Exception
                else:
                    unchecked[a] = 1
        print(p1, p2, sum(unchecked))         
    return count, unchecked
```

```python
def check_triple_primes(count, unchecked, primes, eliminated):
    for i, p1 in enumerate(primes):
        for j, p2 in enumerate(primes[i+1::]):
            for p3 in primes[j + i + 2::]:
                if p1 > 1_000_000/p2/p3:
                    break
                else:
                    to_check = [p1*p2*p3]
                    while(1_000_000/to_check[-1] >= p1):
                        to_check.append(to_check[-1]*p1)
                    stop = False
                    while not stop:
                        stop = True
                        for e in to_check:
                            if 1_000_000/e >= p2:
                                if e * p2 not in to_check:
                                    stop = False
                                    to_check.append(e * p2)
                    stop = False
                    while not stop:
                        stop = True
                        for e in to_check:
                            if 1_000_000/e >= p3:
                                if e * p3 not in to_check:
                                    stop = False
                                    to_check.append(e * p3)
                elims = eliminated[p1].union(eliminated[p2]).union(eliminated[p3])
                count += (1_000_000-len(elims)) * len(to_check)
                for a in to_check:
                    if unchecked[a] == 1:
                        raise Exception
                    else:
                        unchecked[a] = 1
            print(p1, p2, p3, sum(unchecked))  
                   
    return count, unchecked
```

```python
primes = sieveOfAtkin(1_000_000)
eliminated = {p: set([p*i for i in np.arange(1, 1_000_001//p)]) for p in primes}
```

```python
unchecked = [0 for _ in range(1_000_001)]
count = 0
count, unchecked = check_single_primes(count, unchecked, primes, eliminated)
count, unchecked = check_double_primes(count, unchecked, primes, eliminated)
count, unchecked = check_triple_primes(count, unchecked, primes, eliminated)
print(count)
print(sum(unchecked))
```
