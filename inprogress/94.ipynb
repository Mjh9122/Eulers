```python
import math
import numpy as np
```

```python
top = 1_000_000_001//3 + 1

def check_int_area(a, b, c):
    s = (a + b + c)/2
    area = math.sqrt(s * (s-a) * (s-b) * (s-c))
    if type(area) == float:
        if np.round(area) == area and a + b + c < 1_000_000_000:
            print(a, b, c)
            return a + b + c
    return 0

tot = 1
for i in range(2, top):
    if not i % 1_000_000:
        print(100 * i / top)
    a, b, c = i, i, i + 1
    tot += check_int_area(a, b, c)
    c = i - 1
    tot += check_int_area(a, b, c)

print(tot)
```
