```python
with open('../txts/p059_cipher.txt') as cipher_file:
    cipher = cipher_file.readline().strip(" \n").split(',')
    cipher = [int(c) for c in cipher]
print(cipher)
```

```python
for key1 in range(97, 123):
    for key2 in range(97, 123):
        for key3 in range(97, 123):
            message = list()
            for index, char in enumerate(cipher):
                if index % 3 == 0:
                    message.append(chr(char ^ key1))
                if index % 3 == 1:
                    message.append(chr(char ^ key2))
                if index % 3 == 2:
                    message.append(chr(char ^ key3))
            if not any(letter in set(['', '\x7f', "|"]) for letter in message):
                string = ""
                for s in message:
                    string += s
                print(chr(key1), chr(key2), chr(key3), string)
                
```
