```python
def roman_from_num(x):
    roman = ""
    while(x > 0):
        if x >= 1000:
            x -= 1000
            roman += "M"
            continue
        if x >= 900:
            x -= 900
            roman += "CM"
            continue
        if x >= 500 and "D" not in roman:
            x -= 500
            roman += "D"
            continue
        if x >= 400 and "D" not in roman:
            x -= 400
            roman += "CD"
            continue
        if x >= 100:
            x -= 100
            roman += "C"
            continue
        if x >= 90:
            x -= 90
            roman += "XC"
            continue
        if x >= 50 and "L" not in roman:
            x -= 50
            roman += "L"
            continue
        if x >= 40 and "L" not in roman:
            x -= 40
            roman += "XL"
            continue
        if x >= 10:
            x -= 10
            roman += "X"
            continue
        if x >= 9:
            x -= 9
            roman += "IX"
            continue
        if x >= 5 and "V" not in roman:
            x -= 5
            roman += "V"
            continue
        if x >= 4 and "V" not in roman:
            x -= 4
            roman += "IV"
            continue
        if x >= 1:
            x -= 1
            roman += "I"
            continue
    return roman
```

```python
# From stack overflow
def romanToInt(s):
      """
      :type s: str
      :rtype: int
      """
      roman = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000,'IV':4,'IX':9,'XL':40,'XC':90,'CD':400,'CM':900}
      i = 0
      num = 0
      while i < len(s):
         if i+1<len(s) and s[i:i+2] in roman:
            num+=roman[s[i:i+2]]
            i+=2
         else:
            #print(i)
            num+=roman[s[i]]
            i+=1
      return num
```

```python
with open("../txts/p089_roman.txt") as roman_file:
    romans = roman_file.readlines()
count = 0
for r in romans:
    r = r.strip("\n")
    start = len(r)
    num = romanToInt(r)
    end_str = roman_from_num(num)
    end = len(end_str)
    print(end_str)
    if num != romanToInt(end_str):
        print(r, end_str)
    if start > end:
        count += start - end
print(count)
```
