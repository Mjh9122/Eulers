```python
import numpy as np
import math

def step_fac(x):
    str_input = str(x)
    tot = 0
    for i in str_input:
        tot += math.factorial(int(i))
    return int(tot)

known_loops = list()
unique_lengths = dict()

for i in range(1, 1_000_001):
    if not i % 1000:
        print(f"{i/10000} percent done iterating")
    seq = [i]
    for j in range(61):
        next = step_fac(seq[-1])
        b = False
        for loop in known_loops: 
            if next in loop:
                unique_lengths[i] = len(seq) + len(loop)
                b = True
        if b: break
        if next in seq:
            known_loops.append(seq[seq.index(next):])
            unique_lengths[i] = len(seq)
            break
        seq.append(next)

maximum = max(unique_lengths.values())
count = 0
print(f"maximum unique values sequence {maximum}")
for i, j in unique_lengths.items():
    if j == maximum:
        count += 1

print(f"max sequence count {count}")


```
