```python
import numpy as np
import math
import itertools
```

```python
n = 12
arr = [(i, i**2) for i in range(4, 10 ** (n//2) + 1)]



    

def is_s(num, sqr):
    string = str(sqr)
    cuts = [i for i in range(len(string)-1)]
    combs = list()
    for n in range(len(cuts)+1):
        combs += list(itertools.combinations(cuts, n))
    for cut in combs:
        total = 0
        start = 0
        while(start < len(string)):
            end = start
            while end not in cut and end < len(cuts):
                end += 1
            if string[start:end+1] != "":
                total += int(string[start:end+1])
            start = end + 1
        if total == num:
            print(sqr/10**10)
            return True
    return False

print(sum([j for i, j in arr if is_s(i, j)]))
```
